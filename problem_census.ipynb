## Data Preparation

In this step, we will transform the data into the expected format for the algorithm. This step can include some data cleaning, data encoding, data transformations etc.

### Part 1

This dataset hasn’t been cleaned, yet. We can do this by finding all the missing values and handling them.

```python
# TODO
print("Do we have any missing values? ", X_train.isnull().values.any())
print("")

print("Total values: ", len(X_train))
print("Total missing values: ", X_train.isnull().sum().sum())
print("")
print("Total missing values per feature (original): ")
print(X_train.isnull().sum())
print("")

features = list(X_train.columns)
X_train['salary'] = y_train

# We remove missing values in the other 3 features
X_train.dropna(subset = ['workclass'], inplace=True)
X_train.dropna(subset = ['occupation'], inplace=True)
X_train.dropna(subset = ['native-country'], inplace=True)

# We replace missing values in education-num
X_train['education-num'] = X_train['education-num'].fillna(X_train['education-num'].median())

y_train = X_train['salary']
X_train.drop('salary', axis=1, inplace=True)

print("Total values (current): ", len(X_train))
print("Total missing values (current): ", X_train.isnull().sum().sum())
print("")        
print("Total missing values per feature (current): ")
print(X_train.isnull().sum())
print("")
```

### Part 2

All Scikit-learn’s implementations of these algorithms expect numerical features. We check for all features if they are in numerical format. If not, we will transform these features into numerical ones.

```python
# TODO
print("This is our training set at the moment:")
print(X_train)
print("")

# Turn the training set into array
x_train = np.array(X_train)

# Get the unique values from non-numerical features: 'workclass', 'education', 'marital-status' etc.
groups = []
for group in features[3:]:
    groups.append(list(X_train[group].unique()))

from sklearn.preprocessing import OneHotEncoder

# We encoded the non-numerical values using the OneHotEncoder
enc = OneHotEncoder(categories=groups)
encoded_train = enc.fit_transform(x_train[:, 3:]).toarray()

# Finally, the numerical and non-numerical values are joined
x_train = np.concatenate((x_train[:, 0:3], encoded_train), axis=1)

print("This is our training set now (with numerical features):")
print(x_train)
```

### Part 3

```python
# TODO
# We remove the features in test set which are not previously seen
for i in range(0, len(groups)):
    prev_unseen = np.setdiff1d(list(X_test[features[i + 3]].unique()), groups[i])
    if len(prev_unseen) > 0:
        for j in range(0, len(prev_unseen)):
            X_test.drop(X_test.loc[X_test[features[i + 3]] == prev_unseen[j]].index, inplace=True)

# We will also drop the algorithmic bias parts
X_test.drop('race', axis=1, inplace=True)
X_test.drop('sex', axis=1, inplace=True)

# We now have to turn the categorical features into numerical for test set as well
# This is in the same manner as with the training set in the last question
x_test = np.array(X_test)

groups_test = []
for group_t in features[3:]:
    groups_test.append(list(X_test[group_t].unique()))

enc = OneHotEncoder(categories=groups_test)
encoded_test = enc.fit_transform(x_test[:, 3:]).toarray()
x_test = np.concatenate((x_test[:, 0:3], encoded_test), axis=1)
```

## Experiments

### Part 1

Now we set up our experiment. We can divide the data as per the k-fold cross validation method.

```python
# TODO
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate

# We will use K-fold cross-validation
def k_fold_fit_and_evaluate(X, y, model, scoring_method, n_splits=5):
    # define evaluation procedure
    cv = KFold(n_splits=n_splits, random_state=42, shuffle=True)
    # evaluate model
    scores = cross_validate(model, X, y, scoring=scoring_method, cv=cv, n_jobs=-1)
    
    return scores["test_score"]
```

### Part 2

We fit the five algorithms using some default hyper-parameters. We can then create a plot that shows the performances of the algorithms.

```python
# TODO
from sklearn.metrics import f1_score, accuracy_score
from sklearn.metrics import make_scorer

models = {
    "GaussianNB": GaussianNB(),
    "DecisionTreeClassifier": DecisionTreeClassifier(max_depth=None, min_samples_leaf=2, random_state=42),
    "KNeighborsClassifier": KNeighborsClassifier(n_neighbors=3, weights="distance"),
    "SVC": SVC(C=10, kernel="poly", random_state=42),
    "LogisticRegression": LogisticRegression(C=10, penalty='none', random_state=42)
}

X = x_train
y = np.array(y_train)
n_splits = 10
true_target = y

scoring_method = make_scorer(lambda preds, true_target: f1_score(true_target, preds, average="weighted"))

algos = ["GaussianNB", "DecisionTree", "KNN", "SVC", "LogisticReg"]
mean_f1 = []
std_f1 = []

for name, model in models.items():
    print(name)
    metrics_f1 = k_fold_fit_and_evaluate(X, y, model, scoring_method, n_splits=n_splits)
    mean_f1.append(np.mean(metrics_f1))
    std_f1.append(np.std(metrics_f1))
    # START ANSWER
    print("mean F1:", np.mean(metrics_f1))
    print("std F1:", np.std(metrics_f1))
    print("")
    # END ANSWER

plt.scatter(algos, mean_f1)
plt.xlabel("Algorithms")
plt.ylabel("Mean F1")
plt.show()

plt.scatter(algos, std_f1)
plt.xlabel("Algorithms")
plt.ylabel("Std. F1")
plt.show()
```

### Part 3

Now we perform hyper-parameter tuning on the key hyper-parameters previously used.

```python
# TODO
from sklearn.model_selection import GridSearchCV

random_state = 42
n_splits = 10
scoring_method = make_scorer(lambda preds, true_target: f1_score(true_target, preds, average="weighted"))

model_parameters = {
    "GaussianNB": {
        
    },
    "DecisionTreeClassifier": {
        'random_state': [random_state], 
        'max_depth': [None, 2, 5, 10]
    },
    # START ANSWER
    "KNeighborsClassifier": {
        'n_neighbors': list(range(1, 20)), 
        'weights': ["distance"]
    },
    "SVC": {
        'random_state': [random_state],
        'kernel': ["poly"], 
        'C': [0.1, 1, 10, 100]
    },
    "LogisticRegression": {
        'random_state': [random_state],
        'penalty': ["none"], 
        'C': np.logspace(0, 4, 10)
    }
    # END ANSWER
}

tuned_f1 = []
for model_name, parameters in model_parameters.items():
    model = models[model_name]
    
    cv = KFold(n_splits=n_splits, random_state=random_state, shuffle=True)
    grid_search = GridSearchCV(model, parameters, cv=cv, n_jobs=-1, verbose=False, scoring=scoring_method).fit(X, y)
    
    best_model = grid_search.best_estimator_
    best_score = grid_search.best_score_
    best_params = grid_search.best_params_
    tuned_f1.append(best_score)
    
    print(model_name)
    print("- best_score =", best_score)
    print("best parameters:")
    for k,v in best_params.items():
        print("-", k, v)
    print("")
```
